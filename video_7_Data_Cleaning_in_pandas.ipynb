# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Alex course\Pandas Tutorial\Customer Call List(data_cleaning).xlsx")

df
```

```python
#Remove the duplicates
df=df.drop_duplicates() #it will make changes and save bcoz we are storing in df again or else we can use inplace=True
df
```

```python
#remove the unuseful column
df=df.drop(columns='Not_Useful_Column')
df
```

```python
#Cleaning Lastname using strip()
#df['Last_Name'] = df['Last_Name'].str.lstrip("...")
#df['Last_Name'] = df['Last_Name'].str.lstrip("/")
#df['Last_Name'] = df['Last_Name'].str.rstrip("_")
#df

#another way to clean lastname
df['Last_Name'] = df['Last_Name'].str.strip("123._/")  #this removes all at a time only
df
```

```python
#Now phone_number
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True) #replace all value with space other than alphanumeric

df
```

```python
#converting int into string so that we can make a format
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x)) #we can use for loop also
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3]+ '-' + x[3:6] + '-'+ x[6:10])
df
```

```python
#replace na-- from phone_number
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df
```

```python
#Now look at address and clean it 
df['Address'].str.split(',',n=2,expand=True)
```

```python
df[["Street_address","State","Zip_code"]] = df['Address'].str.split(',',n=2,expand=True)
df
```

```python
#Paying_coustomer
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
#Do_not_contact
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
df = df.fillna('')
df
```

```python
#Customer who do not contact we will remove them using the index 
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == "Y":
        df.drop(x,inplace=True)

df
```

```python
#Same for phone number if they are blank do not call them
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x,inplace=True)

df

#another way to drop nullvalues
#df = df.dropna(subset="Phone_Number"),inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```
